**Q1. What is Machine Learning? Explain the main types of Machine Learning with examples.**

Machine Learning (ML) is a branch of Artificial Intelligence that enables computers to learn from data and improve their performance without being explicitly programmed. Instead of following fixed instructions, machine learning systems identify patterns in data and use those patterns to make predictions or decisions.

Machine Learning is widely used in many fields such as healthcare, finance, education, marketing, and real estate.
Types of Machine Learning

Machine Learning is mainly divided into three types:

Supervised Learning
Unsupervised Learning
Reinforcement Learning

**1. Supervised Learning**

Supervised Learning is a type of machine learning in which the model is trained using labeled data. Labeled data means the input data already has the correct output associated with it. The algorithm learns the relationship between input variables and output variables and uses this knowledge to predict results for new data.

Examples

Market basket analysis
Recommendation systems
Common Algorithms

Hierarchical Clustering
Principal Component Analysis (PCA)

Examples
Self-driving cars
Robotics
Game-playing AI
Automated trading systems
Example

A robot learning to walk receives rewards for successful movements and penalties for falling. Over time, it learns the best way to walk.

**Q4. Explain the following evaluation metrics used in Linear Regression:**

**SSE**

**MSE**

**RMSE**

**R² Score**

**Adjusted R²**


In Linear Regression, evaluation metrics are used to measure how well the model predicts the target variable. These metrics help determine the accuracy and performance of the regression model by comparing actual values with predicted values.

The  commonly used evaluation metrics in Linear Regression are:

1. SSE (Sum of Squared Errors)
2. MSE (Mean Squared Error)
3. RMSE (Root Mean Squared Error)
4. R² Score
5. Adjusted R² Score



# 1. SSE (Sum of Squared Errors)

SSE measures the total squared difference between actual values and predicted values.

The formula for SSE is:

SSE = \sum (y_i - \hat{y}_i)^2

Where:

* (y_i) = Actual value
* (\hat{y}_i) = Predicted value



* It calculates the total prediction error in the model.
* Errors are squared to remove negative values.
* Lower SSE indicates better model performance.

## Example

If actual house price is 50 and predicted price is 45:

* Error = 50 − 45 = 5
* Squared error = 25

The model calculates this for all observations and adds them together.



# 2. MSE (Mean Squared Error)

MSE is the average of squared errors.

The formula for MSE is:

MSE = \frac{1}{n}\sum (y_i - \hat{y}_i)^2

Where:

* (n) = Number of observations


* MSE measures the average squared difference between actual and predicted values.
* Lower MSE means better predictions.
* Large errors are penalized more because errors are squared.

## Advantages

* Easy to calculate
* Widely used for model evaluation



# 3. RMSE (Root Mean Squared Error)

RMSE is the square root of MSE.

The formula is:

RMSE = \sqrt{\frac{1}{n}\sum (y_i - \hat{y}_i)^2}



* RMSE measures the average prediction error.
* It is expressed in the same unit as the target variable.
* Lower RMSE indicates better model accuracy.

## Example

If house prices are measured in lakhs, RMSE will also be in lakhs.


RMSE is easier to interpret compared to MSE because it uses original units.



# 4. R² Score (Coefficient of Determination)

R² Score measures how well the independent variables explain the variation in the dependent variable.

The formula is:

R^2 = 1−SSres/SStot

Where:

* (SS_{res}) = Sum of squared residuals
* (SS_{tot}) = Total sum of squares



* R² value ranges from 0 to 1.
* Higher R² indicates better model performance.

## Interpretation

* R² = 0 → Model explains none of the variation
* R² = 1 → Model explains all variation

## Example

If R² = 0.85:

* The model explains 85% of the variation in house prices.



# 5. Adjusted R² Score

Adjusted R² is a modified version of R² that adjusts for the number of independent variables used in the model.

The formula is:

Adjusted\ R^2 = 1−(n−p−1(1−R2)(n−1)​)

Where:

* (n) = Number of observations
* (p) = Number of predictors



* Adjusted R² increases only if the new variable improves the model significantly.
* It prevents unnecessary variables from making the model appear better.



* More reliable than R² when multiple predictors are used.

In the house price prediction problem, the workflow helps in preparing the data, training the model, and predicting house prices effectively.

1. Data Collection
2. Data Preprocessing
3. Exploratory Data Analysis (EDA)
4. Feature Selection
5. Splitting the Dataset
6. Model Building
7. Model Training
8. Model Evaluation

**Q10. Train a Linear Regression model and perform the following:**

**Make predictions**

**Calculate MAE**

**Calculate MSE**

**Calculate RMSE**

**Calculate R² Score**

**Calculate Adjusted R²**

**Write the Linear Regression equation**

**Interpret the model coefficients**

````python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
````

````python
model =LinearRegression()
model.fit(X_train,y_train)
````

````python
y_pred =model.predict(X_test)
print(y_pred)
````

````python
mae = mean_absolute_error(y_test,y_pred)
print("MAE:", mae)
````

````python
mse = mean_squared_error(y_test, y_pred)

print("MSE:", mse)
````

````python
rmse = np.sqrt(mse)

print("RMSE:", rmse)
````

````python
r2 = r2_score(y_test, y_pred)

print("R² Score:", r2)
````

````python
n = X_test.shape[0]   # Number of observations
p = X_test.shape[1]   # Number of features

adjusted_r2 = 1 - ((1 - r2) * (n - 1) / (n - p - 1))

print("Adjusted R²:", adjusted_r2)
````

````python
print("Intercept:", model.intercept_)

print("Coefficients:", model.coef_)
````

````python
coefficients = pd.DataFrame({
    'Feature': X.columns,
    'Coefficient': model.coef_
})

print(coefficients)
````

The coefficients of the Linear Regression model indicate the impact of each feature on house price (medv). Positive coefficients increase house price, while negative coefficients decrease house price. The variable rm has a strong positive effect on house prices, whereas nox, ptratio, and lstat negatively affect house prices. Among all features, pollution level (nox) shows the strongest negative impact on house price.
